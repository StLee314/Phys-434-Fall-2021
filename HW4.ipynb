```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

### Problem 0

```python
t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10 + np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
```

### Problem 1

### 1A:

```python
# Generating scatterplot of first 1000 points
scatterPlot = t[0:1000]
x = bkgd[0:1000]
fig, ax = plt.subplots(1, 1, figsize=(10,10))
ax.set_title("First 1k Points in Background Data", fontsize=18)
ax.set_xlabel("t", fontsize=16)
ax.set_ylabel("Background", fontsize=16)
plt.scatter(scatterPlot, x)
plt.tick_params(labelsize = 16)
plt.show()
```

The first 1000 data points of the given data set from Problem 0 are plotted on a scatterplot.

### 1B:

```python
def create_scatter_plot(t, background, start, stop, step) :
    new_t = t[start:stop:step]
    new_background = background[start:stop:step]
    scatterPlot = t[0:1000]
    fig, ax = plt.subplots(1, 1, figsize=(10,10))
    ax.set_title("Graph from Point " + str(start) + " to " + str(stop) + " with Steps of " + str(step), fontsize=20)
    ax.set_xlabel("t", fontsize=16)
    ax.set_ylabel("Background", fontsize=16)
    plt.scatter(new_t, new_background)
    plt.tick_params(labelsize = 16)
    plt.show()
```

create_scatter_plot allows us to chop off and graph specific sections from the data set depending on the beginning and ending values. We also have the ability to ommit certain data points by only plotting certain values with the step variable.

```python
create_scatter_plot(t, bkgd, 1500, 4700, 1)
```

### 1C:

```python
create_scatter_plot(t, bkgd, 0, 100000, 100)
```

By setting the step variable to 100, we can plot every 100th data point in the data set.

### Problem 2

```python
bin_size = 100
fig, ax = plt.subplots(1, 1, figsize=(10,10))
ax.set_title("2D Histogram Showing the Central Density of a Data Set", fontsize = 16)
ax.set_xlabel("t", fontsize=16)
ax.set_ylabel("Background", fontsize=16)
plt.tick_params(labelsize = 16)
plt.hist2d(t, bkgd, density = True, bins = bin_size)
plt.colorbar()
plt.show()
```

### 2A:

Above is a 2D histogram showing the central density of the data points for each value of t, against the corresponding background distribution that was given in the beginning of the assignment. 

### 2B:

Here, the color of each point in the histogram signifies the density of points in each bin from each array in the x-axis and y-axis. Yellow signifies that there is a high density of points, and purple means that there is a lower density of points.

```python
t_folded = t % (2*np.pi) #creating new folded time variable
create_scatter_plot(t_folded, bkgd, 0, 100000, 1)
```
